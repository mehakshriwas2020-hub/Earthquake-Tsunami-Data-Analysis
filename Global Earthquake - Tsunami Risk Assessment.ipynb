```python
# Group earthquake counts per year
earthquake_count = df.groupby('Year').size()

# Group tsunami events per year
tsunami_count = df.groupby('Year')['tsunami'].sum()

plt.figure(figsize=(10,5))

plt.plot(earthquake_count.index, earthquake_count.values,
         marker='<', linestyle='-', label='Total Earthquakes')

plt.plot(tsunami_count.index, tsunami_count.values,
         marker='o', linestyle='--', label='Tsunami Events')

plt.title("Earthquake & Tsunami Trends (2001‚Äì2022)")
plt.xlabel("Year")
plt.ylabel("Number of Events")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by Year to calculate average magnitude
yearly_magnitude = df.groupby("Year")["magnitude"].mean().reset_index()

# Plot the average magnitude trend
plt.figure(figsize=(12,6))
plt.plot(yearly_magnitude["Year"], yearly_magnitude["magnitude"],
         marker='o', color='purple')

plt.title("Average Earthquake Magnitude (2001‚Äì2022)")
plt.xlabel("Year")
plt.ylabel("Average Magnitude")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Magnitude distribution
plt.subplot(1,2,1)
sns.histplot(df['magnitude'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')

# Depth distribution
plt.subplot(1,2,2)
sns.histplot(df['depth'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Earthquake Depths')
plt.xlabel('Depth (km)')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group by tsunami (0 = No, 1 = Yes) and calculate mean
avg_stats = df.groupby('tsunami')[['magnitude','depth']].mean().reset_index()
avg_stats['Tsunami'] = avg_stats['tsunami'].map({0:'No', 1:'Yes'})

# Plotting
plt.figure(figsize=(8,4))
bar_width = 0.35
x = range(len(avg_stats))

# Average Magnitude
plt.bar(x, avg_stats['magnitude'], width=bar_width, label='Average Magnitude', color='green')

# Average Depth
plt.bar([i + bar_width for i in x], avg_stats['depth'],
        width=bar_width, label='Average Depth', color='brown')

plt.xticks([i + bar_width/2 for i in x], avg_stats['Tsunami'])
plt.ylabel('Value')
plt.title('Average Magnitude and Depth: Tsunami vs Non-Tsunami')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Earthquake_data_tsunami (2).csv")
```

```python
# Convert time column to datetime and extract year if not already present
if "Year" not in df.columns:
    df["time"] = pd.to_datetime(df["time"], errors="coerce")
    df["Year"] = df["time"].dt.year

# Filter major earthquakes (magnitude ‚â• 8.0)
major_eq = df[df["magnitude"] >= 8.0]
print(f"Total Major Earthquakes (‚â•8.0): {len(major_eq)}\n")

#  Summary Statistics
print("Summary Statistics for Major Earthquakes:\n")
summary = major_eq[["magnitude", "depth", "latitude", "longitude", "Year", "tsunami"]].describe()
print(summary)

# Geographic Distribution Plot
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=major_eq,
    x="longitude", y="latitude",
    size="magnitude", hue="tsunami",
    sizes=(100,300), palette="coolwarm", alpha=0.7
)
plt.title("Geographic Distribution of Major Earthquakes (‚â•8.0)", fontsize=13)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Tsunami (1=Yes, 0=No)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# Yearly Frequency Trend
plt.figure(figsize=(10,6))
sns.countplot(x="Year", data=major_eq, color="orange")
plt.title("Number of Major Earthquakes (‚â•8.0) by Year (2001‚Äì2022)", fontsize=13)
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# Magnitude vs Depth Visualization
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=major_eq,
    x="magnitude", y="depth",
    hue="tsunami", palette="coolwarm", s=100, alpha=0.8
)
plt.title("Magnitude vs Depth of Major Earthquakes (‚â•8.0)", fontsize=13)
plt.xlabel("Magnitude")
plt.ylabel("Depth (km)")
plt.legend(title="Tsunami (1=Yes, 0=No)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'],
            c=df['magnitude'], cmap='Reds', edgecolors='k')

plt.title('Geographic Distribution of Earthquakes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Magnitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Non-Tsunami
plt.scatter(df[df['tsunami'] == 0]['longitude'],
            df[df['tsunami'] == 0]['latitude'],
            c='blue', label='Non-Tsunami')

# Tsunami
plt.scatter(df[df['tsunami'] == 1]['longitude'],
            df[df['tsunami'] == 1]['latitude'],
            c='red', label='Tsunami')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x=df['longitude'],
    y=df['latitude'],
    color='green',
    s=30,
    alpha=0.6,
    label='Tsunami Events',
    edgecolor='k'
)

plt.title('Density of Tsunami Events (Hotspot Visualization)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='tsunami', y='magnitude',  hue='tsunami',  palette=['blue', 'red'])
plt.xticks([0, 1], ['Non-Tsunami', 'Tsunami'])
plt.title('Magnitude Comparison between Tsunami and Non-Tsunami Events')
plt.xlabel('')
plt.ylabel('Magnitude')
plt.grid(True, alpha=0.3)
plt.show()

plt.figure(figsize=(10,6))
sns.histplot(df[df['tsunami']==0]['magnitude'], bins=10, color='blue', label='Non-Tsunami', alpha=0.6)
sns.histplot(df[df['tsunami']==1]['magnitude'], bins=10, color='red', label='Tsunami', alpha=0.6)
plt.title('Magnitude Distribution: Tsunami vs Non-Tsunami')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()

avg_values = df.groupby('tsunami')[['magnitude','depth','sig']].mean().reset_index()
avg_values['tsunami'] = avg_values['tsunami'].map({0: 'Non-Tsunami', 1: 'Tsunami'})

avg_values.plot(x='tsunami', kind='bar', figsize=(8,5), colormap='coolwarm')
plt.title('Average Seismic Feature Comparison')
plt.ylabel('Average Value')
plt.xlabel('')
plt.show()
```

```python
df['mag_range'] = pd.cut(df['magnitude'],
                         bins=[0,5,6,7,8,10],
                         labels=['<5','5-6','6-7','7-8','8+'])

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='mag_range', hue='tsunami')

plt.title('Tsunami Occurrence by Magnitude Range')
plt.xlabel('Magnitude Range')
plt.ylabel('Count')
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(8,6))
corr = df[['magnitude','depth','sig','tsunami']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Seismic Features')
plt.show()
```

## Conclusion
